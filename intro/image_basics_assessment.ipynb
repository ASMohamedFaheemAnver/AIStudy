```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("dog_backpack.jpg")
```

```python
plt.imshow(img)
```

```python
fixed_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fixed_img)
```

```python
fliped_img = cv2.flip(fixed_img, 0)
```

```python
plt.imshow(fliped_img)
```

```python
rec_img = cv2.rectangle(fixed_img.copy(), (200, 350), (600, 700), (255, 0, 0), 10)
```

```python
plt.imshow(rec_img)
```

```python
vertices = np.array([[380, 280], [600, 700], [200, 700]], np.int32)
```

```python
vertices.shape
```

```python
pts = vertices.reshape(-1, 1, 2)
```

```python
pts.shape
```

```python
pts
```

```python
rec_img = cv2.fillPoly(fixed_img.copy(), [pts], color=(255, 0, 0))
```

```python
plt.imshow(rec_img)
```

```python
def create_circle(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img, (x, y), 100, (0, 0, 255), thickness=10)
        
img = fixed_img
cv2.namedWindow("Dog")
cv2.setMouseCallback("Dog", create_circle)

while True:
    cv2.imshow("Dog", img)
    if cv2.waitKey(20) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()
```
