```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def display_img(img, cmap=None):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
giaraffe_img = plt.imread("giraffes.jpg")
```

```python
display_img(giaraffe_img)
giaraffe_img.shape
```

```python
gray_giaraffe = cv2.imread("giraffes.jpg", 0)
```

```python
ret, thresh1 = cv2.threshold(gray_giaraffe, 127, 255, cv2.THRESH_BINARY)
```

```python
display_img(thresh1, "gray")
```

```python
img = cv2.imread("giraffes.jpg")
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
display_img(img)
```
